```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import loader as ld   #permet de charcher les données et d'avoir la matrice user/items
import show as sh     #permet d'affichier des matrices
import matrix as mmat #permet d'effectuer des opérations sur les matrices
              
#########################################################################################################################

#Compte le nombre d'users et d'items
users, items = ld.check_data("ml-100k/u.data")

#Construction matrice user/items
mat_u_i =ld.matrice_user_item("ml-100k/u.data", users, items)

#Différentes approximations par moyennes
mat_cu = ld.remplissage_moyenne_user(mat_u_i)
mat_cf = ld.remplissage_moyenne_film(mat_u_i)

#Affichage des 3 matrices
sh = sh.show()
sh.draw_mat([ mat_u_i, mat_cu, mat_cf])
```

```python
#Approximation bas rang avec k = 100
k = 30

Dec_user = mmat.simplifier(mat_cu,k)
Dec_film = mmat.simplifier(mat_cf,k)

M1 = np.dot( Dec_user[0], np.dot(Dec_user[1],Dec_user[2]))
M2 = np.dot( Dec_film[0], np.dot(Dec_film[1],Dec_film[2]))

sh.draw_mat([M1,M2])
```
